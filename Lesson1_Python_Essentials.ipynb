### 1.3 Variables and Data Types
Variables store data. Python is dynamically typed, so you don’t need to declare the type explicitly.

```python
# Integer
number_of_elements = 3
print(number_of_elements)

# Float (decimal)
band_gap = 2.5
print(band_gap)

# String
material_name = "TiO2"
print(material_name)

# Boolean
is_metal = False
print(is_metal)
```

**Common data types**:
- `int` – whole numbers (e.g., `5`, `-2`, `1000`)
- `float` – decimal numbers (e.g., `3.14`, `2.0`, `-0.5`)
- `str` – text (e.g., `"Ag"`, `'Cu'`, `"ZnO"`)
- `bool` – `True` or `False`

### 1.4 Basic Operators
- Arithmetic: `+`, `-`, `*`, `/`, `//` (integer division), `%` (modulo), `**` (power).
- Comparison: `==`, `!=`, `<`, `>`, `<=`, `>=` → return a boolean.
- Logical: `and`, `or`, `not`.

**Example**:

```python
a = 10
b = 3
print(a + b)   # 13
print(a / b)   # 3.333...
print(a // b)  # 3
print(a % b)   # 1
print(a ** b)  # 1000

print(a > b)   # True
print(a == b)  # False
```

## 2. Data Structures for Materials Data

### 2.1 Lists
A list is an ordered, mutable collection of items. Use square brackets `[]`.

```python
elements = ["Ag", "Cu", "Zn", "Ti"]
print(elements)
print(elements[0])      # first element (index 0)
print(elements[-1])     # last element
print(elements[1:3])    # slice: elements at index 1 and 2

elements.append("Au")   # add an element
print(elements)

elements.remove("Cu")   # remove by value
print(elements)

len(elements)           # number of items
```

### 2.2 Tuples
A tuple is an ordered, **immutable** collection. Use parentheses `()`.

```python
atomic_numbers = (47, 29, 30, 22)   # Ag, Cu, Zn, Ti
print(atomic_numbers[2])             # 30
# atomic_numbers[2] = 31             # This would cause an error (immutable)
```

Use tuples when you have data that should not change (e.g., coordinates, constants).

### 2.3 Dictionaries
A dictionary stores key‑value pairs. Use curly braces `{}`.

```python
# Dictionary of element symbols to atomic numbers
atomic_dict = {
    "Ag": 47,
    "Cu": 29,
    "Zn": 30,
    "Ti": 22
}
print(atomic_dict["Ag"])        # 47

# Add a new entry
atomic_dict["Au"] = 79
print(atomic_dict)

# Check if a key exists
if "Fe" in atomic_dict:
    print(atomic_dict["Fe"])
else:
    print("Fe not found")
```

Dictionaries are extremely useful for mapping properties to element symbols or material IDs.

### 2.4 Sets
A set is an unordered collection of unique items. Use `set()` or curly braces without colons.

```python
unique_elements = {"Ag", "Cu", "Zn", "Ag", "Ti"}
print(unique_elements)        # duplicates removed: {'Ag', 'Cu', 'Zn', 'Ti'}

unique_elements.add("Au")
print(unique_elements)
```

Sets are handy for removing duplicates or performing mathematical set operations (union, intersection).

## 3. Control Flow

### 3.1 Conditional Statements (if‑elif‑else)

```python
band_gap = 2.5

if band_gap < 1.0:
    print("Metal or narrow band gap")
elif band_gap < 3.0:
    print("Semiconductor")
else:
    print("Insulator")
```

### 3.2 Loops

**`for` loop** – iterate over a sequence.

```python
elements = ["Ag", "Cu", "Zn"]
for el in elements:
    print(f"Element: {el}")

# Using range
for i in range(5):          # 0,1,2,3,4
    print(i)
```

**`while` loop** – repeat while a condition is true.

```python
count = 0
while count < 3:
    print(count)
    count += 1
```

## 4. Functions

Functions encapsulate reusable code.

```python
def classify_material(band_gap):
    """Return a string classification based on band gap."""
    if band_gap < 1.0:
        return "metal"
    elif band_gap < 3.0:
        return "semiconductor"
    else:
        return "insulator"

# Use the function
gap = 2.5
print(classify_material(gap))      # semiconductor
```

**Components of a function**:
- `def` keyword
- function name (e.g., `classify_material`)
- parameters (`band_gap`)
- docstring (optional but recommended)
- `return` statement (optional)

## 6. Putting It All Together: A Nanoinformatics Mini‑Example

Let’s combine what we’ve learned to perform a simple task: read a list of material formulas, count the occurrence of each formula, and compute the average number of elements per formula.

```python
# List of material formulas (simplified)
formulas = ["Ag", "CuO", "ZnO", "TiO2", "Ag", "CuO"]

# Count occurrences of each formula
from collections import Counter
counts = Counter(formulas)
print("Formula counts:", counts)

# Extract all element symbols from the formulas
all_elements = []
for f in formulas:
    # A very simple parser: just split into characters (not chemically accurate!)
    # In real life, you'd use pymatgen.
    for ch in f:
        if ch.isalpha():
            all_elements.append(ch)

print("All elements:", all_elements)
print("Unique elements:", set(all_elements))
print("Number of elements per formula:", [len(f) for f in formulas])
```
