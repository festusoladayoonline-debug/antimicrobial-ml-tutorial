## 1. Introduction to NumPy

### 1.1 What is NumPy?
NumPy (Numerical Python) is the fundamental package for scientific computing in Python. It provides:
- A powerful N-dimensional array object (`ndarray`)
- Fast vectorised operations (no explicit loops needed)
- Tools for linear algebra, Fourier transforms, and random number generation
- Integration with C/C++ and Fortran code

In nanoinformatics, NumPy is used extensively for handling large arrays of numerical data, such as band gaps, formation energies, and atomic coordinates.

### 1.3 Creating NumPy Arrays
The main object in NumPy is the homogeneous multidimensional array (`ndarray`). Let's create arrays from Python lists:

```python
import numpy as np

# Create a 1D array from a list
arr1 = np.array([1, 2, 3, 4, 5])
print("1D array:", arr1)
print("Type:", type(arr1))
print("Shape:", arr1.shape)
print("Data type:", arr1.dtype)

# Create a 2D array (matrix)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D array:\n", arr2)
print("Shape:", arr2.shape)  # (rows, columns)

# Create arrays with built-in functions
zeros = np.zeros((3, 4))      # 3x4 array of zeros
ones = np.ones((2, 3))        # 2x3 array of ones
identity = np.eye(3)          # 3x3 identity matrix
random_arr = np.random.rand(2, 2)  # 2x2 random numbers between 0 and 1

print("\nZeros:\n", zeros)
print("\nOnes:\n", ones)
print("\nIdentity:\n", identity)
print("\nRandom:\n", random_arr)
```

### 1.4 Array Indexing and Slicing
NumPy arrays support indexing and slicing similar to Python lists, but with additional capabilities for multidimensional arrays.

```python
arr = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print("Original array:\n", arr)

# Indexing
print("Element at row 0, col 1:", arr[0, 1])   # 20
print("First row:", arr[0, :])                 # [10 20 30]
print("Second column:", arr[:, 1])             # [20 50 80]

# Slicing
print("Submatrix (first 2 rows, last 2 columns):\n", arr[:2, 1:3])

# Boolean indexing (select elements > 50)
print("Elements > 50:", arr[arr > 50])
```

### 1.5 Vectorised Operations
One of NumPy's greatest strengths is the ability to perform element-wise operations on entire arrays without explicit loops. This is called **vectorisation** and leads to much faster code.

```python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

# Element-wise operations
print("a + b =", a + b)
print("a * b =", a * b)
print("a ** 2 =", a ** 2)
print("np.sqrt(a) =", np.sqrt(a))
print("np.sin(a) =", np.sin(a))

# Comparisons return boolean arrays
print("a > 2 =", a > 2)
print("a == 3 =", a == 3)
```

### 1.6 Useful NumPy Functions
NumPy provides a wide range of functions for statistical and mathematical operations.

```python
data = np.array([2.5, 3.1, 4.0, 2.8, 3.6])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard deviation:", np.std(data))
print("Min:", np.min(data))
print("Max:", np.max(data))
print("Sum:", np.sum(data))
print("Cumulative sum:", np.cumsum(data))
```

## 2. Introduction to Pandas

### 2.1 What is Pandas?
Pandas is a powerful library for data manipulation and analysis. It introduces two main data structures:
- **Series**: one-dimensional labeled array (like a column in a spreadsheet)
- **DataFrame**: two-dimensional labeled data structure (like a table)

Pandas is built on top of NumPy and is the tool of choice for working with tabular data in Python. In this course, we will use Pandas to handle the results of Materials Project queries, engineer features, and prepare data for machine learning.

```python
import pandas as pd

# Creating a Series from a list
s = pd.Series([1, 3, 5, 7, 9], name="odd_numbers")
print("Series:\n", s)
print("Index:", s.index)
print("Values:", s.values)

# Creating a DataFrame from a dictionary
data = {
    "Element": ["Ag", "Cu", "Zn", "Ti"],
    "Atomic_Number": [47, 29, 30, 22],
    "Atomic_Mass": [107.87, 63.55, 65.38, 47.87],
    "Band_Gap": [0.0, 0.0, 3.3, 0.0]  # Approximate band gaps in eV
}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)
```

### 2.7 Handling Missing Data
Real-world datasets often have missing values. Pandas provides tools to detect and handle them.

```python
# Create a DataFrame with missing values
df_missing = pd.DataFrame({
    "A": [1, 2, np.nan, 4],
    "B": [5, np.nan, 7, 8],
    "C": [9, 10, 11, 12]
})
print("Original DataFrame with missing values:\n", df_missing)

# Check for missing values
print("\nIs null?\n", df_missing.isnull())
print("\nCount of nulls per column:\n", df_missing.isnull().sum())

# Drop rows with any missing values
df_dropped = df_missing.dropna()
print("\nAfter dropping rows with any NaN:\n", df_dropped)

# Fill missing values with a constant
df_filled = df_missing.fillna(0)
print("\nAfter filling with 0:\n", df_filled)

# Fill missing values with the mean of the column
df_filled_mean = df_missing.fillna(df_missing.mean())
print("\nAfter filling with column means:\n", df_filled_mean)
```

## 3. Data Aggregation and Grouping

### 3.1 GroupBy Operations
The `groupby` method allows you to split data into groups based on some criteria, apply a function to each group, and combine the results.

```python
# Import pandas first
import pandas as pd

# Create a larger DataFrame
data = {
    "Material": ["Ag", "Ag", "Cu", "Cu", "Zn", "Zn"],
    "Crystal_System": ["cubic", "cubic", "cubic", "cubic", "hexagonal", "hexagonal"],
    "Band_Gap": [0.0, 0.0, 0.0, 0.0, 3.3, 3.3],
    "Formation_Energy": [-0.1, -0.12, -0.15, -0.14, -0.2, -0.22]
}
df_materials = pd.DataFrame(data)
print("Materials DataFrame:\n", df_materials)

# Group by Material and compute mean of numeric columns
grouped_mean = df_materials.groupby("Material").mean(numeric_only=True)
print("\nMean by Material:\n", grouped_mean)

# Group by Crystal_System and compute multiple statistics using named aggregation
grouped_stats = df_materials.groupby("Crystal_System").agg(
    Band_Gap_Mean=('Band_Gap', 'mean'),
    Band_Gap_Std=('Band_Gap', 'std'),
    Formation_Energy_Min=('Formation_Energy', 'min'),
    Formation_Energy_Max=('Formation_Energy', 'max')
)
print("\nStatistics by Crystal System:\n", grouped_stats)
```

## 4. Merging and Joining DataFrames

Often you need to combine data from multiple sources. Pandas provides several functions for this: `merge()`, `join()`, and `concat()`.

```python
# Create two DataFrames
df1 = pd.DataFrame({
    "Element": ["Ag", "Cu", "Zn"],
    "Atomic_Radius": [144, 128, 134]  # pm
})

df2 = pd.DataFrame({
    "Element": ["Ag", "Cu", "Zn"],
    "Electronegativity": [1.93, 1.90, 1.65]
})

# Merge on the common column 'Element'
merged = pd.merge(df1, df2, on="Element")
print("Merged DataFrame:\n", merged)

# Concatenate along rows
df3 = pd.DataFrame({
    "Element": ["Ti", "Au"],
    "Atomic_Radius": [147, 144],
    "Electronegativity": [1.54, 2.54]
})
concatenated = pd.concat([merged, df3], ignore_index=True)
print("\nConcatenated DataFrame:\n", concatenated)
```

## 5. Working with Realistic Nanoinformatics Data

Now let's apply what we've learned to a more realistic dataset. We'll create a DataFrame that mimics the structure of data we might get from the Materials Project.

```python
# Simulate Materials Project data
np.random.seed(42)
n_materials = 20

materials_data = {
    "material_id": [f"mp-{i}" for i in range(1000, 1000+n_materials)],
    "formula": np.random.choice(["Ag", "Cu", "Zn", "Ti", "Au", "Mg", "Fe"], n_materials),
    "band_gap": np.random.uniform(0, 5, n_materials),
    "formation_energy_per_atom": np.random.uniform(-2, 1, n_materials),
    "density": np.random.uniform(2, 12, n_materials),
    "volume": np.random.uniform(50, 200, n_materials)
}

df_mp = pd.DataFrame(materials_data)
print("Simulated Materials Project DataFrame:")
df_mp.head(10)
```

### 5.1 Basic Data Exploration

```python
print("DataFrame info:")
df_mp.info()

print("\nSummary statistics:")
df_mp.describe()

print("\nUnique formulas:", df_mp["formula"].unique())
print("\nCount of each formula:")
print(df_mp["formula"].value_counts())
```

### 5.2 Filtering and Querying

```python
# Find materials with band gap < 3.0 (potential semiconductors)
semiconductors = df_mp[df_mp["band_gap"] < 3.0]
print("Potential semiconductors (band gap < 3.0 eV):")
print(semiconductors[["formula", "band_gap"]])

# Find materials with formation energy < -1.0 (thermodynamically stable)
stable = df_mp[df_mp["formation_energy_per_atom"] < -1.0]
print("\nThermodynamically stable materials (formation energy < -1.0 eV/atom):")
print(stable[["formula", "formation_energy_per_atom"]])
```

### 5.3 Adding Engineered Features
In later lessons, we will use `pymatgen` to extract detailed elemental properties. For now, let's add some simple features manually.

```python
# Create a dictionary of atomic masses (simplified)
atomic_mass = {
    "Ag": 107.87,
    "Cu": 63.55,
    "Zn": 65.38,
    "Ti": 47.87,
    "Au": 196.97,
    "Mg": 24.31,
    "Fe": 55.85
}

# Map atomic mass to each row
df_mp["atomic_mass"] = df_mp["formula"].map(atomic_mass)

# Create a simple feature: mass_density_ratio
df_mp["mass_density_ratio"] = df_mp["atomic_mass"] / df_mp["density"]

print("DataFrame with engineered features:")
df_mp.head()
```

### 5.4 Grouping and Aggregating

```python
# Average properties by element
grouped = df_mp.groupby("formula").agg({
    "band_gap": ["mean", "std"],
    "formation_energy_per_atom": "mean",
    "density": "mean"
})
print("Average properties by element:")
print(grouped.round(3))
```
